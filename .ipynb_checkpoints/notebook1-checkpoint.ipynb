```python
import pandas as pd

imd = pd.read_csv('data/imd_2019.csv', encoding='utf-8-sig')
print(imd.shape)
print(imd.columns.tolist())
print(imd.head(3))
```

```python
import geopandas as gpd

gdf = gpd.read_file('data/lsoa_boundaries.geojson')
print(gdf.shape)
print(gdf.columns.tolist())
print(gdf.head(3))
```

```python
IMD_LSOA_COL = 'LSOA11CD'
IMD_SCORE_COL = 'IMD19'
GEO_LSOA_COL = 'LSOA11CD'

merged = gdf.merge(imd[[IMD_LSOA_COL, IMD_SCORE_COL]], 
                   left_on=GEO_LSOA_COL, 
                   right_on=IMD_LSOA_COL, 
                   how='inner')

print(f"Merged rows: {len(merged)}")
print(merged[[GEO_LSOA_COL, IMD_SCORE_COL]].head())
```

```python
merged = merged.to_crs('EPSG:4326')

# Calculate centroids correctly
merged['centroid'] = merged.geometry.to_crs('EPSG:27700').centroid.to_crs('EPSG:4326')
merged['lat'] = merged['centroid'].y
merged['lon'] = merged['centroid'].x

# Filter to London using lat/lon bounding box instead
london = merged[
    (merged['lat'] > 51.28) & (merged['lat'] < 51.70) &
    (merged['lon'] > -0.51) & (merged['lon'] < 0.33)
]

print(f"London LSOAs: {len(london)}")

# Pick 5 most deprived and 5 least deprived
most_deprived = london.nsmallest(5, IMD_SCORE_COL)
least_deprived = london.nlargest(5, IMD_SCORE_COL)

test_lsoas = pd.concat([most_deprived, least_deprived])
test_lsoas = test_lsoas[[GEO_LSOA_COL, IMD_SCORE_COL, 'lat', 'lon']].reset_index(drop=True)

print(test_lsoas)
test_lsoas.to_csv('data/test_lsoas.csv', index=False)
print("Saved!")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 8))
london.plot(ax=ax, color='lightgrey', edgecolor='white', linewidth=0.3)
most_deprived.plot(ax=ax, color='red', markersize=50, label='Most deprived', zorder=5)
least_deprived.plot(ax=ax, color='green', markersize=50, label='Least deprived', zorder=5)
ax.set_title('Your 10 Test LSOAs in London')
ax.axis('off')
plt.legend()
plt.savefig('outputs/test_lsoas_map.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
london['decile'] = pd.qcut(london[IMD_SCORE_COL], q=10, labels=False)
sample = london.groupby('decile').apply(
    lambda x: x.sample(50, random_state=42)
).reset_index(drop=True)
sample = sample[[GEO_LSOA_COL, IMD_SCORE_COL, 'lat', 'lon']]

print(f"Total LSOAs sampled: {len(sample)}")
print(sample.head())
sample.to_csv('data/test_lsoas.csv', index=False)
print("Saved!")
```
